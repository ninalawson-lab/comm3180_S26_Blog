#### Happiest Countries + USA

```python
happiest = (
    big5_scored_df.query('country in ["US","FI","DK","IS","SE","IL"]').
        groupby('country')[['O','C','E','A','N']].
        agg('mean').
        round(1).
        transpose().
        plot(kind='barh')
)
```
